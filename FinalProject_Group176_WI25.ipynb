## Background and Prior Work


It has long been believed that physical activity plays an important role in the general health condition of an individual, with extensive research continuously being done on a wide range of exercise and their effect on health conditions. One of the most common and universally accessible physical activities one can engage in is the simple act of walking, which can be conveniently tracked with digital devices such as smartphones and is often measured in daily steps. Meanwhile, sleep quality issues, such as sleep disorder like sleep apnea as well as general difficulty in sleeping, is commonly believed to be correlated with physical activities <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). More specifically, it is commonly believed that increased levels of physical activity generally improves sleep quality and reduces the risk of sleep disorders <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). 

Our project aims to explore this topic further, with a specific focus on the effect of daily step count on the sleep quality of middle-aged overweight individuals. We believe that this is an interesting and meaningful topic of exploration since bad sleep quality is often perceived to be more pervasive among middle-aged individuals, while the overweight trait could often be an indication for a lack of physical activity, which can potentially highlight the effect of different levels of physical activity with more clarity.

As per prior studies, there is a generally positive correlation between exercise and reduced sleeping issues. By exercising, many experience more time in bed actually sleeping, reduced sleepiness during the day, and an overall higher quality of life <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Similarly, weight loss associated with physical activity can contribute to better quality sleep as well <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). 

However, despite an extensive amount of research on sleep quality, most of the studies rely heavily on subjective metrics such as survey responses rather than more objective metrics, such as step count, when measuring physical activity level <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Moreover, even the studies that do use more quantifiable metrics such as step counts often do so in self report which once again pose the risk of inaccurate data. Most studies also tend to target a diverse population with less emphasis on specific audiences <a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). While most studies suggest a generally positive correlation between physical activity and sleep quality with little conflicting findings, the lack of specificity poses the risk of hidden cofounders influencing the reliability of findings. These factors together motivate our project to focus on a specific, high-risk audience using a clear, quantitative metric to measure physical activity, with the potential of identifying biases and filling gaps in current works.

### Data Wrangling

Here, we will do some data wrangling before our initial exploration. In particular, we will first want to narrow down the variables to include only those that we are interested in. In particular, we want to keep the variables bmi, steps, age, gender, and sleep quality variables including minutesAsleep, sleep_deep_ratio, sleep_wake_ratio, sleep_light_ratio and sleep_rem_ratio. We would also want to keep id and date for now, as we need these identifiers for further wrangling. We might come back to look at some of the removed variables later, but most of them are either unrelated to our topic, or is collected through self reports and are thus not considered for our project.

```python
# filter out the target columns and assign to a new df
filterd_df = raw_df[['id', 'date', 'bmi', 'steps', 'gender', 'age', 'minutesAsleep', 'sleep_deep_ratio', 'sleep_wake_ratio', 'sleep_light_ratio', 'sleep_rem_ratio']]
filterd_df
```

```python
# Check out entries with a specific id, representing a single participant's statistics on different days
filterd_df[filterd_df['id'] == '621e339967b776a240e502de']
```

Immediately we can see many NaN values in our data, this is because some of the participants recorded in the dataset participated through surveys rather then tracking through smartwatch, and not all participants wears the smartwatch during their sleep. In some cases, data that shouldn't change drastically during the periods in the records, such as gender and bmi, were simply missing. While missing sleep and step count data cannot be filled in arbitrarily, we can fill in some of the missing bmi and gender. In addition, we want to ensure that no entry with the same id and date exists, since that will be a duplicate.

```python
# Attempt to find duplicate entries
duplicates = filterd_df[filterd_df.duplicated(subset=['id', 'date'], keep=False)]
duplicates
```

```python
# No duplicates found, proceed to generate a list of all unique id
unique_ids = filterd_df["id"].unique().tolist()
# Total number of unique participants
print(len(unique_ids))
# Note while we have many observations with detailed information, the actual pool of participants is actually rather small. This is a potential limitation to our project
# For each unique id, attempt to find a non-NaN bmi value and fill in any NaN bmi value for entries with the same id. The dataset contains data collected in 4 month period,
# we are assuming that bmi for the same individual would not vary drastically during this timeframe.

# First, look at the list of unique bmi values
print(filterd_df['bmi'].value_counts())
# Check unique types in the BMI column
print(filterd_df["bmi"].apply(type).value_counts())
# Seems bmi values are either float or string, we can remove entries with bmi values lower then 25 since we only considers overweight individuals

# Function to convert BMI values
def parse_bmi(v):
    if isinstance(v, str):
        v = v.strip()  # remove space
        if v.startswith(">="):  
            return float(v[2:])  # extract from ">=xx"
        elif v.startswith("<"):  
            return float(v[1:]) - 0.1  # convert "<xx" to just below xx, will remove anyway
        try:
            return float(v)  # convert normal numeric strings
        except ValueError:
            return np.nan
    elif isinstance(v, (int, float)):  
        return float(v)  # no change
    return np.nan  # all non-convertible values is NaN

# add new bmi_numeric column
filterd_df["bmi_numeric"] = filterd_df["bmi"].apply(parse_bmi)
# make sure new column is only numerics and NaN
print(filterd_df["bmi_numeric"].unique())
# Filter based on BMI over 25, also keep NaN
filterd_df_overweight = filterd_df[(filterd_df["bmi_numeric"] >= 25) | (filterd_df["bmi_numeric"].isna())]
```

```python
# New unique BMI values in the filtered df
print(filterd_df_overweight['bmi'].value_counts())

# Find the first non-NaN bmi_numeric value for each unique id
bmi_by_id = filterd_df_overweight.groupby("id")["bmi_numeric"].transform(lambda x: x.ffill().bfill())

# Fill NaN values in bmi_numeric using the found values
filterd_df_overweight["bmi_numeric"] = filterd_df_overweight["bmi_numeric"].fillna(bmi_by_id)

# Check is still have NaN value
print(filterd_df_overweight[filterd_df_overweight["bmi_numeric"].isna()])  # Should be empty if all NaNs are filled

# Verify if any id has more than one unique BMI after filling (should be 1 per id)
print(filterd_df_overweight.groupby("id")["bmi_numeric"].nunique().max())  # expect 1
```

The NaN values should belong to individuals that either have no reported BMI at all, or have reported BMI that is not overweight and is removed earlier, hence leaving entries that have no found BMI to fill. These entries can then be removed completely.

```python
# Drop rows where bmi_numeric is still NaN
filterd_df_overweight = filterd_df_overweight.dropna(subset=["bmi_numeric"])
# value counts unique BMI values in the filtered df, should not have changed
print(filterd_df_overweight['bmi'].value_counts())
# Check is still have NaN value
print(filterd_df_overweight[filterd_df_overweight["bmi_numeric"].isna()])  # Should be empty if all NaNs are filled
# Take a look at the newest data
filterd_df_overweight
```

Notice that there are still NaN in the bmi category, but all bmi_numeric is filled. The NaN in bmi can then be safely ignored since we wont need explicit bmi values for now, and the dateset is correctly sorted to only known overweight entries. We will now continue to filter age.

```python
# Take a look at age of the remaining entries
print(filterd_df_overweight['age'].value_counts())
# Take a look at how many NaN is in age
print(len(filterd_df_overweight[filterd_df_overweight['age'].isna()]))  # Should be 0 if no NaN

# Remove entries aged below 30
filterd_df_overweight_middleAged = filterd_df_overweight[filterd_df_overweight["age"] != "<30"]

# Function to convert age values
def parse_age(v):
    if isinstance(v, str) and v.startswith(">="):  
        return float(v[2:])  # Extract number from ">=30"
    return np.nan  # Keep NaN values as is

# Convert age to age_numeric
filterd_df_overweight_middleAged["age_numeric"] = filterd_df_overweight_middleAged["age"].apply(parse_age)

# Fill NaN age with any age found for same id
filterd_df_overweight_middleAged["age_numeric"] = filterd_df_overweight_middleAged.groupby("id")["age_numeric"].transform(lambda x: x.ffill().bfill())
filterd_df_overweight_middleAged = filterd_df_overweight_middleAged.dropna(subset=["age_numeric"])  # Drop entries still NaN
print(len(filterd_df_overweight_middleAged[filterd_df_overweight_middleAged['age'].isna()]))  # Should be 0 if no NaN

# Take a look at current df
filterd_df_overweight_middleAged
```

Now all data remaining should have a bmi and age that is within our range of interest. However we still have NaN in many steps count entries and sleep-related variables. For now, we will remove all entries with a NaN in step count and minutesAsleep and put the result in a new dataframe. However, we will keep the current dataframe so we can come back later and potentially analyze these entries with self-reported subjective metrics that we have currently excluded, to provide more comparison.

```python
# How many entries have empty step count?
print(len(filterd_df_overweight_middleAged[filterd_df_overweight_middleAged['steps'].isna()]))
filterd_df_overweight_middleAged_dropNa = filterd_df_overweight_middleAged.dropna(subset=["steps"])  # Drop entries with NaN in steps
print(len(filterd_df_overweight_middleAged_dropNa[filterd_df_overweight_middleAged_dropNa['steps'].isna()]))    # should be 0

# How many entries have empty minutesAsleep?
print(len(filterd_df_overweight_middleAged[filterd_df_overweight_middleAged['minutesAsleep'].isna()]))
filterd_df_overweight_middleAged_dropNa = filterd_df_overweight_middleAged.dropna(subset=["minutesAsleep"])  # Drop entries with NaN in minutesAsleep
print(len(filterd_df_overweight_middleAged_dropNa[filterd_df_overweight_middleAged_dropNa['minutesAsleep'].isna()]))    # should be 0

# Newest dataframe
filterd_df_overweight_middleAged_dropNa
```

```python
# Take a look at any remaining NaN values
print(filterd_df_overweight_middleAged_dropNa.isna().sum())
```

We now have a filtered out dataframe that contains entries of individuals in the overweight category, aged over 30 that we have step count and sleep metrics for. Note that we have some remaining NaN values in more specific sleep metrics other than minutesAsleep, but we will keep them in for now to do some exploration just between step counts and sleep duration. We will remove them later when more specific analysis with these metrics needs to be done. This is a rather small subset in comparison to the original, and we might want to explore different settings and conditions later, but as of now this is a good starting point for some EDA.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches #idea from stackoverflow.com
```

```python
#Visualize relationship between step count and minutes asleep -- Scatter Plot

#Set figure size
plt.figure(figsize=(10,8)) 

#Create scatter plot with regression line
sns.regplot(x='steps', y='minutesAsleep', data=filterd_df_overweight_middleAged_dropNa, color='red')

#Labels for x-axis, y-axis, and title
plt.xlabel('Daily Step Count')
plt.ylabel('Sleep Duration (minutes)')
plt.title('Relationship Between Step Count and Sleep Duration')
```

This plot represents the relationship between the sleep duration and the daily step count of the individuals in the study. Clearly, the data is scattered without a clear relationship between the two. However, there is a cluster around approximately 380 minutes of sleep, which translates to 6 hours and 20 minutes and about 11,000 steps per day. Overall though, there is no distinct relationship.

```python
#Visualize relationship between step count and minutes asleep -- Contour Density Plot

#Set figure size
plt.figure(figsize=(10,8))

#Creating a contour plot
sns.kdeplot(x='steps', y='minutesAsleep', data=filterd_df_overweight_middleAged_dropNa, color='red')

#Labels for x-axis, y-axis, and title
plt.xlabel('Daily Step Count')
plt.ylabel('Sleep Duration (minutes)')
plt.title('Relationship Between Step Count and Sleep Duration')
```

### Statistical tests

We will perform statistical tests below to further analyze the dataset using an alternative from visualization:

```python
from scipy.stats import pearsonr, ttest_ind

# Calculate Pearson correlation coefficient (r-value) and p-value for each sleep metric
def calculate_correlation(df, x_col, y_col):
    # Drop rows with NaN values in the columns of interest
    df_cleaned = df.dropna(subset=[x_col, y_col])
    r_value, p_value = pearsonr(df_cleaned[x_col], df_cleaned[y_col])
    return r_value, p_value

# Step count vs. Sleep duration
r_steps_sleep, p_steps_sleep = calculate_correlation(filterd_df_overweight_middleAged, 'steps', 'minutesAsleep')
print(f"Step Count vs. Sleep Duration: r = {r_steps_sleep:.3f}, p = {p_steps_sleep:.3f}")

# Step count vs. Deep sleep ratio
r_steps_deep, p_steps_deep = calculate_correlation(sleep_ratio_filterd_df, 'steps', 'sleep_deep_ratio')
print(f"Step Count vs. Deep Sleep Ratio: r = {r_steps_deep:.3f}, p = {p_steps_deep:.3f}")

# Step count vs. Wake time ratio
r_steps_wake, p_steps_wake = calculate_correlation(sleep_ratio_filterd_df, 'steps', 'sleep_wake_ratio')
print(f"Step Count vs. Wake Time Ratio: r = {r_steps_wake:.3f}, p = {p_steps_wake:.3f}")

# Step count vs. Light sleep ratio
r_steps_light, p_steps_light = calculate_correlation(sleep_ratio_filterd_df, 'steps', 'sleep_light_ratio')
print(f"Step Count vs. Light Sleep Ratio: r = {r_steps_light:.3f}, p = {p_steps_light:.3f}")

# Step count vs. REM sleep ratio
r_steps_rem, p_steps_rem = calculate_correlation(sleep_ratio_filterd_df, 'steps', 'sleep_rem_ratio')
print(f"Step Count vs. REM Sleep Ratio: r = {r_steps_rem:.3f}, p = {p_steps_rem:.3f}")
```

As can be seen from the result above, The correlation do seem to suggest that an increase in step count reduces wake time and light sleep ratio and increases sleep duration, REM sleep and deep sleep ratio. However, all of the r value are low in magnitude and the p values were all very high, suggesting that the correlations observed are not statistically significant.

### Sleep metrics transformation
Our current correlation seems rather unclear when using raw values from the sleep ratio metrics. Below, we will transform the sleep metrics into categorical data and generate boxplots and violin plots to see if we can observe clearer correlation.

```python
sleep_filterd_df_transform_1 = sleep_ratio_filterd_df
# Function to categorize sleep ratios into "Decreased", "Normal", and "Increased"
def categorize_sleep_ratio(ratio):
    if ratio <= 0.95:
        return "Decreased"
    elif ratio >= 1.05:
        return "Increased"
    else:
        return "Normal"

sleep_ratio_cols = ["sleep_deep_ratio", "sleep_light_ratio", "sleep_rem_ratio", "sleep_wake_ratio"]

# Apply categorization to each sleep ratio column and create new categorical columns
for col in sleep_ratio_cols:
    sleep_filterd_df_transform_1[col + "_category"] = sleep_filterd_df_transform_1[col].apply(categorize_sleep_ratio)

# Function to plot boxplots and violin plots for step count vs categorical sleep metrics
def plot_step_count_vs_sleep_category(df, sleep_category_col, title):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Boxplot
    sns.boxplot(x=sleep_category_col, y="steps", data=df, order=["Decreased", "Normal", "Increased"], ax=axes[0])
    axes[0].set_title(f"Boxplot: {title}", fontsize=14)
    axes[0].set_xlabel(sleep_category_col.replace("_", " ").title(), fontsize=12)
    axes[0].set_ylabel("Step Count", fontsize=12)

    # Violin plot
    sns.violinplot(x=sleep_category_col, y="steps", data=df, order=["Decreased", "Normal", "Increased"], ax=axes[1])
    axes[1].set_title(f"Violin Plot: {title}", fontsize=14)
    axes[1].set_xlabel(sleep_category_col.replace("_", " ").title(), fontsize=12)
    axes[1].set_ylabel("Step Count", fontsize=12)

    plt.tight_layout()
    plt.show()

# Generate plots for each sleep category
for sleep_col in ["sleep_deep_ratio_category", "sleep_light_ratio_category", "sleep_rem_ratio_category", "sleep_wake_ratio_category"]:
    plot_step_count_vs_sleep_category(sleep_filterd_df_transform_1, sleep_col, f"Step Count vs {sleep_col.replace('_', ' ')}")
```

We also attempt to calculate the t value and r value again, this time using categorical data by converting 'Decreased' category to -1, 'Normal' Category to 0, and 'Increased' category to 1, in the hope of reducing the influence of extreme outliers and high fluctuation in raw values and instead to observe the general trend more clearly.

```python
# Encoding_map, suggested by chatGPT
encoding_map = {"Decreased": -1, "Normal": 0, "Increased": 1}

for col in ["sleep_deep_ratio_category", "sleep_light_ratio_category", "sleep_rem_ratio_category", "sleep_wake_ratio_category"]:
    sleep_filterd_df_transform_1[col + "_encoded"] = sleep_filterd_df_transform_1[col].map(encoding_map)

# Step count vs. Deep sleep ratio
r_steps_deep, p_steps_deep = calculate_correlation(sleep_filterd_df_transform_1, 'steps', 'sleep_deep_ratio_category_encoded')
print(f"Step Count vs. Deep Sleep Ratio: r = {r_steps_deep:.3f}, p = {p_steps_deep:.3f}")

# Step count vs. Wake time ratio
r_steps_wake, p_steps_wake = calculate_correlation(sleep_filterd_df_transform_1, 'steps', 'sleep_wake_ratio_category_encoded')
print(f"Step Count vs. Wake Time Ratio: r = {r_steps_wake:.3f}, p = {p_steps_wake:.3f}")

# Step count vs. Light sleep ratio
r_steps_light, p_steps_light = calculate_correlation(sleep_filterd_df_transform_1, 'steps', 'sleep_light_ratio_category_encoded')
print(f"Step Count vs. Light Sleep Ratio: r = {r_steps_light:.3f}, p = {p_steps_light:.3f}")

# Step count vs. REM sleep ratio
r_steps_rem, p_steps_rem = calculate_correlation(sleep_filterd_df_transform_1, 'steps', 'sleep_rem_ratio_category_encoded')
print(f"Step Count vs. REM Sleep Ratio: r = {r_steps_rem:.3f}, p = {p_steps_rem:.3f}")
```

As can be seen from the graphs and the statistical tests, although a slightly clearer correlation can be seen using the categorical metrics, the overall correlation remains unclear and the p values are very high for the Deep Sleep Ratio and Wake Time Ratio. However, the statistical significance between correlation of step count and REM sleep and light sleep, did see a significant increase when using categorical data. However, the magnitude of r value remains minimal. We will continue to take a brief look at our second dataset on the correlation between step count and sleep disorder as a simple complement.

# Ethics & Privacy

Biases that can be raised with our research question include selection bias, reporting bias, confounding variables, and recall bias. Reporting bias and recall bias would lead to misinformation due to false information from overreporting or underreporting if the information is being volunteered from the participant. Selection bias and confounding variables would lead to misrepresentation of the results from the study. This may include factors such as socioeconomic status, physical health, medications, and more. As our research question investigates health concerns and habits, privacy concerns consist of HIPAA compliance, informed consent, and anonymity or de-identification of the participants in the study. Terms of use issues would concern how the data is collected and shared. Potential biases in our dataset may include selection bias regarding factors such as over or underrepresentation of certain groups including certain geographic locations, disabilities, or socioeconomic groups. These factors may be problematic in terms of it allowing for equitable analysis.

We will set out to detect reporting bias and recall bias before communicating our analysis by identifying how the data was collected, whether it was self reported in a survey or if the results came from a source other than the participant recording the information. Selection bias will be detected through anonymity of the participant and pre-survey before data is collected, verifying they qualify for the preliminary requirements of the study to reduce confounding factors such as pre-existing health concerns.

Some other issues that may be potentially problematic in terms of data privacy and equitable impact may be that if we collect data through the usage of wearable fitness trackers, then that could exclude individuals who do not have access to that technology or individuals who cannot afford said devices. Consequently, the dataset might be biased in terms of underrepresenting individuals from lower-income backgrounds or ethnic minorities that don’t have access to the fitness trackers or health services. Furthermore, the dataset being collected contains sensitive health-related information, such as sleep disorders, BMI, sleep patterns, etc. As a result, it is necessary that the data is handled with caution in order to preserve the privacy of the individuals. Improper usage or access to the data will result in breaches of privacy and confidentiality. 

In order to handle the issues that we have identified above, we will ensure that any data collected follows privacy standards such as HIPAA and that the data will be anonymized to protect the privacy of the individuals. To have equity in the representation of the dataset, we will use diverse data sources that will include individuals from multiple demographic backgrounds. We will also use a large enough sample size to account for any possible differences between the groups. For the individuals who may not have access to fitness trackers or the technology, we can consider methods of including said individuals or we can analyze only those with similar access to technology and acknowledge the limitations of the dataset availability.
